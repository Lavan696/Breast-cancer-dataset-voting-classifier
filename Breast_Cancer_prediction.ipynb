```python
corr_with_target = X.corrwith(y)
corr_with_target.sort_values(ascending=False).plot(kind='bar', figsize=(10,4))
plt.title('Correlation of Features with Target')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
X_scaled = std_scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X_scaled,y , test_size = 0.2,random_state = 42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_features': ['sqrt', 'log2'],
    'max_depth' : [7,8],
    'criterion' :['gini', 'entropy'],
    'max_leaf_nodes' :[16]

}
grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=StratifiedKFold(n_splits = 10),verbose = 2, n_jobs = -1)
```

```python
grid_search_rf.fit(X_train,y_train)
```

```python
grid_search_rf.best_params_
```

```python
rf_best_estimator = grid_search_rf.best_estimator_
```

```python
importances = rf_best_estimator.feature_importances_
```

```python
importances = pd.Series(importances*100, index=X.columns).sort_values(ascending = False)
importances
```

```python
top_n = 10
top_features = importances.head(top_n).index.tolist()
print(f"Top {top_n} features used for retraining:\n", top_features)
```

```python
top_features_importance = importances.head(top_n)
plt.figure(figsize=(10, 6))

sns.barplot(
    x=top_features_importance.values,
    y=top_features_importance.index,
    hue=top_features_importance.index,
    palette='crest',
    dodge=False,
    legend=False
)

plt.title('Top 20 Most Important Features (Random Forest Classifier)')
plt.xlabel('Importance Score---->')
plt.ylabel('Feature---->')
plt.show()
```

```python
from sklearn.svm import LinearSVC
param_grid_lsvc = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'max_iter': [1000, 2000, 5000]
}
grid_search_lsvc = GridSearchCV(LinearSVC(random_state=42), param_grid_lsvc, cv=StratifiedKFold(n_splits=10), verbose=2, n_jobs=-1)
```

```python
grid_search_lsvc.fit(X_train,y_train)
```

```python
grid_search_lsvc.best_params_
```

```python
from sklearn.linear_model import LogisticRegression
param_grid_lr = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'max_iter': [1000, 2000, 5000]
}
grid_search_lr = GridSearchCV(LogisticRegression(random_state=42), param_grid_lr, cv=StratifiedKFold(n_splits=10), verbose=2, n_jobs=-1)
```

```python
grid_search_lr.fit(X_train, y_train)
```

```python
grid_search_lr.best_params_
```

```python
from sklearn.ensemble import StackingClassifier
stacking_clf = StackingClassifier(
    estimators = [
        ('lr' , grid_search_lr.best_estimator_),
        ('rf', grid_search_rf.best_estimator_),
        ('lsvc',grid_search_lsvc.best_estimator_)
     ],
    final_estimator = grid_search_rf.best_estimator_

)
```

```python
stacking_clf.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
stacking_clf_cvs = cross_val_score(stacking_clf,X_train,y_train,cv = 10,scoring = 'accuracy')
stacking_clf_cvs
```

```python
stacking_clf_cvs.mean()
```

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    stacking_clf, X_train, y_train, cv=5, scoring='accuracy'
)
train_mean = train_scores.mean(axis=1)
test_mean = test_scores.mean(axis=1)

plt.figure(figsize=(8,6))
plt.plot(train_sizes, train_mean, 'o-', label='Training Score')
plt.plot(train_sizes, test_mean, 'o-', label='Cross-validation Score')
plt.xlabel('Training Examples')
plt.ylabel('Accuracy')
plt.title('Learning Curve - Stacking Classifier')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_probabilities = stacking_clf.predict_proba(X_test)
y_probabilities
```

```python
y_predicted = stacking_clf.predict(X_test)
y_predicted
```

```python
from sklearn.calibration import calibration_curve
prob_true, prob_pred = calibration_curve(y_test, y_probabilities[:,1], n_bins=10)
plt.plot(prob_pred, prob_true, marker='o')
plt.title('Calibration Curve')
plt.xlabel('Predicted Probability')
plt.ylabel('True Probability')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cmd = ConfusionMatrixDisplay(cm,display_labels= data.target_names)
cmd.plot(cmap = plt.cm.Greens )
plt.title('Confusion Matrix for Stacking Classifier')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr,threshold = roc_curve(y_test,y_probabilities[:,1])
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([-0.05, 1.05])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic(ROC)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_test,y_probabilities[:,1])
```

```python
plt.plot(thresholds, precisions[:-1] , 'b--', label = 'Precision', linewidth = 2)
plt.plot(thresholds, recalls[:-1],'g-', label = 'Recall', linewidth = 2)
plt.xlabel('Thresholds')
plt.title('Precison Recall Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_predicted, output_dict=True)
clf_report = pd.DataFrame(clf_report).transpose()
clf_report = clf_report.drop(columns=['support'])
plt.figure(figsize=(8, 6))
sns.heatmap(clf_report.iloc[:-3, :].T, annot=True, cmap='BuGn', fmt=".3f")
plt.title('Classification Report Heatmap')
plt.ylabel('Metrics--->')
plt.xlabel('Classes--->')
plt.show()
```
