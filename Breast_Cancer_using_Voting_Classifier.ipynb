```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer(as_frame = True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y , test_size = 0.2,random_state = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
voting_clf = VotingClassifier(
    estimators = [
        ('lr' , LogisticRegression(random_state = 42,max_iter = 5000)),
        ('rf', RandomForestClassifier(random_state = 42,max_leaf_nodes= 16)),
        ('svc',SVC(random_state = 42,probability= True))   
     ]
   
)
voting_clf.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
cross_score = cross_val_score(voting_clf,X_train,y_train,cv = 5,scoring = 'accuracy')
cross_score
```

```python
cross_score.mean()
```

```python
y_predicted = voting_clf.predict(X_test)
```

```python
voting_clf.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predicted)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cmd = ConfusionMatrixDisplay(cm,display_labels= voting_clf.classes_)
cmd.plot(cmap = plt.cm.Greens )
plt.title('Confusion Matrix for Voting Classifier')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score
precision_score(y_test,y_predicted)
```

```python
recall_score(y_test,y_predicted)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_predicted)
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_test,y_predicted)
```

```python
plt.plot(thresholds, precisions[:-1] , 'b--', label = 'Precision', linewidth = 2)
plt.plot(thresholds, recalls[:-1],'g-', label = 'Recall', linewidth = 2)
plt.xlabel('Thresholds')
plt.title('Precison Recall Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_predicted)
```
